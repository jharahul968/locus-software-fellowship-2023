```python
import pandas as pd
import numpy as np
import sklearn as skt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("Linear Regression - Sheet1.csv")
data=data.reset_index(drop=True)
data=data.iloc[:298,:]
```

```python
data
```

```python
x=data['X'].to_numpy()
```

```python
y=data['Y'].to_numpy()
```

```python
plt.plot(x,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
def liner_function(m,x,c):
  return m*x + c
```

```python
epochs=1000
n=X_train.shape[0]
L=0.001
m=0
c=0
predicted=[]
train_losses=[]
test_losses=[]
le=1e-10
for i in range(epochs):
     Y_pred = liner_function(m,X_train,c)
     Y_test_pred=liner_function(m,X_test,c)
     loss=(y_train - Y_pred)**2
     train_loss=loss[0]
     test_loss=(y_test-Y_test_pred)**2
     test_loss_app=test_loss[0]
     D_m = (-2/n) * sum(X_train * (y_train - Y_pred))
     D_c = (-2/n) * sum(y_train - Y_pred)
     m = m - L * D_m+le  # Update m
     c = c - L * D_c+le  # Update c
     predicted.append(Y_pred)
     train_losses.append(train_loss)
     test_losses.append(test_loss_app)
```

```python
X=8
Y_pred = m*X + c
```

```python
print(Y_pred)
```

```python
print(m)
print(c)
```

```python
plt.plot(train_losses,label="Train loss")
plt.plot(test_losses,label="Test loss")
plt.legend()
plt.show()
```

```python
plt.plot(X_train,predicted[0],label="1 iteration")
plt.plot(X_train,predicted[50],label="50 iteration")
plt.plot(X_train,predicted[100],label="100 iteration")
plt.plot(X_train,predicted[200],label="200 iteration")
plt.plot(X_train,predicted[999],label="1000 iteration")
plt.legend()
plt.show()
```
